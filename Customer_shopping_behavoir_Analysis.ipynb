```python
# Checking for null values
df.isnull().sum()
```

```python
#Impute null values using fillna median strategy by grouping category column
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
# Checking for null values
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

**Feature Engineering**

```python
# Create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)      # qcut() - Quantile Cut. It divides the Age values into equal-sized groups based on the number of rows.
```

```python
df[['age','age_group']].head(5)
```

```python
# Create a new column frequency_purchase_days using frequency_of_purchases
frequency_mapping = {
    'weekly': 7,
    'fortnightly': 14,
    'bi-weekly': 14,
    'monthly': 30,
    'quarterly': 90,
    'every 3 months': 90,
    'annually': 365
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].str.lower().map(frequency_mapping)
```

```python
df[['frequency_of_purchases','purchase_frequency_days']].head(5)
```
